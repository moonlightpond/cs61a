```python
def count_stair_ways(n):
    if n == 0:
        return 1
    elif n < 0:
        return 0
    else:
        return count_stair_ways(n-1) + count_stair_ways(n-2)
```

# 1.2

count_k is different from count_partitions. \
To count_k, order matters. To count different partitions, order doesn't matter.

```python
def count_k(n, k):
    """
    >>> count_k(3, 3) # 3, 2 + 1, 1 + 2, 1 + 1 + 1
    4
    >>> count_k(4, 4)
    8
    >>> count_k(10, 3)
    274
    >>> count_k(300, 1) # Only one step at a time
    1
    """
    if k == 0 or n < 0:
        return 0
    elif n == 0:
        return 1
    else:
        i = 1 
        total = 0
        while i <= k:
            total += count_k(n-i, k)
            i += 1
        return total
    
count_k(4, 4)
```

```python
def even_weighted(s):
    """
    >>> x = [1, 2, 3, 4, 5, 6]
    >>> even_weighted(x)
    [0, 6, 20]
    """
    return [i * s[i] for i in range(len(s)) if i % 2 == 0]

x = [1, 2, 3, 4, 5, 6]
even_weighted(x)
```

```python
def max_product(s):
    """Return the maximum product that can be formed using non-consecutive
    elements of s.
    >>> max_product([10,3,1,9,2]) # 10 * 9
    90
    >>> max_product([5,10,5,10,5]) # 5 * 5 * 5
    125
    >>> max_product([])
    1
    """
    def helper(index, product):
        if index >= len(s):
            return product
        else:
            return max(helper(index + 1, product), helper(index + 2, product * s[index]))
    return helper(0, 1)
```

```python
def max_product2(s):
    """Return the maximum product that can be formed using non-consecutive
    elements of s.
    >>> max_product([10,3,1,9,2]) # 10 * 9
    90
    >>> max_product([5,10,5,10,5]) # 5 * 5 * 5
    125
    >>> max_product([])
    1
    """
    if len(s) == 0:
        return 1
    elif len(s) == 1:
        return s[0]
    else:
        return max(max_product2(s[1:]), s[0] * max_product2(s[2:]))
```
