```python
def multiply(m, n):
    """
    >>> multiply(5, 3)
    15
    """
    if n == 1:
        return m
    else: 
        return m + multiply(m, n-1)
    
multiply(5, 6)
```

```python
def hailstone(n):
    """Print out the hailstone sequence starting at n, and return the
    number of elements in the sequence.
    >>> a = hailstone(10)
    10
    5
    16
    8
    4
    2
    1
    >>> a
    7
    """
    print(n)
    if n == 1:
        return 1
    elif n%2 == 0:
        return 1 + hailstone(n/2)
    else:
        return 1 + hailstone(n*3+1)
    
a = hailstone(10)
a
```

```python
def merge(n1, n2):
    """ Merges two numbers
    >>> merge(31, 42)
    4321
    >>> merge(21, 0)
    21
    >>> merge (21, 31)
    3211
    """
    def helper(k, a1, a2, result):    
        if a1 == 0:
            return a2
        elif a2 == 0:
            return a1
        elif a1 < 10:
            return (max(a1, a2)*10 + min(a1, a2)) * 10**k + result
        else:
            return helper(k+2, a1//10, a2//10, (max(a1%10, a2%10)*10 + min(a1%10, a2%10)) * 10**k + result)         
    return helper(0, n1, n2, 0)

merge(31, 42)
merge(21, 31)
```

```python
def make_func_repeater(f, x):
    """
    >>> incr_1 = make_func_repeater(lambda x: x + 1, 1)
    >>> incr_1(2) #same as f(f(x))
    3
    >>> incr_1(5)
    6
    """
    def repeat(n):
        if n == 1:
            return f(x)
        else:
            return f(repeat(n-1))
    return repeat

incr_1 = make_func_repeater(lambda x: x + 1, 1)
incr_1(2)
incr_1(5)
```

```python
def is_prime(n):
    """
    >>> is_prime(7)
    True
    >>> is_prime(10)
    False
    >>> is_prime(1)
    False
    """
    def prime_helper(k):
        if n == k:
            return True
        elif n == 1 or n % k == 0:
            return False
        else:
            return prime_helper(k+1)
    return prime_helper(2)

is_prime(6)
```
