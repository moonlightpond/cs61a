# 1.7 Recursive Function

### A function is recursive if it calls itslef in its body.

```python
def sum_digits(x):
    """ return the sum of the digits of a positive integer x. """
    assert(x > 0)
    assert(type(x) == int)
    
    if x < 10:
        return x
    else:
        all_but_last, last = x // 10, x % 10
        return sum_digits(all_but_last) + last
    return 1

sum_digits(123)
```

```python
def fac(n):
    """ return the factorial of a positive integer n. """
    if n == 1:
        return 1
    else:
        return n * fac(n - 1)
    
fac(3)  
```

### If dividing a recursive procedure into two functions that call each other, then the two functions are mutually recursively.

```python
def is_odd(n):
    """ n has to be positive integers """
    if n == 1:
        return True
    elif n == 0:
        return False
    else:
        return is_even(n-1)
    
def is_even(n):
    """ n has to be positive integers """
    if n == 0:
        return True
    elif n == 0:
        return False
    else:
        return is_odd(n-1)
    
is_even(6)
```

```python
# a and b take turns to remove pebbles from n inital pebbles. 
# a always removes one.
# b removes one if n is odd and two otherwise.
def play_a(n):
    if n == 0:
        print("a wins!")
    else:
        play_b(n-1)

def play_b(n):
    if n == 0:
        print("b wins!")
    elif is_odd(n):
        play_a(n-1)
    else:
        play_a(n-2)
        
play_a(20)
```

### A function is tree recursive if it calls itself more than once

```python
def fib(n):
    """ Calculate Fibonacci numbers, in which each number is the sum of the preceeding two """
    if n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return fib(n-1) + fib(n-2)
    
fib(3)
```

```python
def count_partitions(n, m):
    """ Return the number of different partitions of n using intergers up to m"""
    if n == 0:
        return 1
    elif n < 0:
        return 0
    elif m == 0:
        return 0
    else:
        return count_partitions(n-m, m) + count_partitions(n, m-1)
    
count_partitions(4, 2)
```
